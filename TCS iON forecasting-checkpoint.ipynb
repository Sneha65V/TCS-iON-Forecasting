#### Exploratory Data Analysis

```python
# plotting sales w.r.t. item
plt.figure(figsize=(10,4))
plt.title("Items Vs Sales")
plt.xticks(rotation=90)
sns.barplot(x=data['item'],y=data['sales']);
```

```python
# plotting sales w.r.t. item
plt.figure(figsize=(10,4))
plt.title("Store Vs Sales")
sns.barplot(x=data['store'],y=data['sales']);
```

```python
# plotting correlation matrix
corr_matrix = data.corr()
plt.title('Correlation Between different columns')
sns.heatmap(corr_matrix,annot=True,xticklabels=True, yticklabels=True,linewidth=.5,cmap='Spectral')
plt.show()
```

No high correlation is observed between any columns. That means that the variables are independent of each other

```python
# Getting the sales of particular items
item_data=data.groupby(['item']).agg({'sales':['min','max','count']})
item_data
```

```python
# Getting the sales of particular store
store_data=data.groupby(['store']).agg({'sales':['min','max','count']})
store_data
```
